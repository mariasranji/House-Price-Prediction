#### Importing 

```python
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
import lightgbm as lgb
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.ensemble import GradientBoostingRegressor
from xgboost import XGBRegressor
from sklearn.preprocessing import PolynomialFeatures
from sklearn.impute import SimpleImputer
```

#### X, Y split

```python
X = df.drop("SalePrice", axis=1)
y = df["SalePrice"]
```

#### One-hot Encoding

```python
X = pd.get_dummies(X, columns=cat_cols)
```

#### Standardizing the data

```python
important_num_cols.remove("SalePrice")

scaler = StandardScaler()
X[important_num_cols] = scaler.fit_transform(X[important_num_cols])
```

```python
X.head()
```

#### Train - Test Split

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=121)
def rmse_cv(model):
    rmse = np.sqrt(-cross_val_score(model, X, y, scoring="neg_mean_squared_error", cv=5)).mean()
    return rmse
    

def evaluation(y, predictions):
    mae = mean_absolute_error(y, predictions)
    mse = mean_squared_error(y, predictions)
    rmse = np.sqrt(mean_squared_error(y, predictions))
    r_squared = r2_score(y, predictions)
    return mae, mse, rmse, r_squared

models = pd.DataFrame(columns=["Model","MAE","MSE","RMSE","R2 Score","RMSE (Cross-Validation)"])
```

##### Linear Regression

```python
l_reg = LinearRegression()
l_reg.fit(X_train, y_train)
predictions = l_reg.predict(X_test)
```

```python
predictions
```

```python

mae, mse, rmse, r_squared = evaluation(y_test, predictions)
print("MAE:", mae)
print("MSE:", mse)
print("RMSE:", rmse)
print("R2 Score:", r_squared)
print("-"*30)
rmse_cross_val = rmse_cv(l_reg)
print("RMSE Cross-Validation:", rmse_cross_val)

new_row = pd.DataFrame([{
    "Model": "LinearRegression",
    "MAE": mae,
    "MSE": mse,
    "RMSE": rmse,
    "R2 Score": r_squared,
    "RMSE (Cross-Validation)": rmse_cross_val
}])

models = pd.concat([models, new_row], ignore_index=True)
```

### Saving the Trained Model and Preprocessing Steps
To deploy this model or use it for future predictions, we need to save the model along with the preprocessing states (the `StandardScaler` and the exact columns expected after `pd.get_dummies`). We will use `joblib` to bundle and save them.

```python
import joblib

# Create a model bundle containing the model and all preprocessing components
model_bundle = {
    'model': gb_reg_full,
    'scaler': scaler,
    'important_num_cols': important_num_cols,
    'cat_cols': cat_cols,
    'model_columns': X.columns.tolist()
}

# Save the bundle to a file
joblib.dump(model_bundle, 'gradient_boosting_house_price_model.pkl')
print("Best model (GradientBoostingRegressor) and preprocessing steps saved successfully!")
```

### How to Load and Use the Saved Model on New Data
Here is the code to load the artifact and correctly preprocess a new dataset before making predictions.

```python
# Load the saved bundle
loaded_bundle = joblib.load('gradient_boosting_house_price_model.pkl')
loaded_model = loaded_bundle['model']
loaded_scaler = loaded_bundle['scaler']
saved_num_cols = loaded_bundle['important_num_cols']
saved_cat_cols = loaded_bundle['cat_cols']
model_columns = loaded_bundle['model_columns']

def predict_new_data(new_df):
    """
    Preprocesses new data identically to the training data and returns predictions.
    """
    # 1. Select the relevant features
    df_features = new_df[saved_num_cols + saved_cat_cols].copy()
    
    # 2. Perform One-Hot Encoding for categorical features
    df_features = pd.get_dummies(df_features, columns=saved_cat_cols)
    
    # 3. Align dummy columns with training format (fills missing categories with 0, drops extras)
    df_features = df_features.reindex(columns=model_columns, fill_value=0)
    
    # 4. Scale numerical columns using the saved scaler
    df_features[saved_num_cols] = loaded_scaler.transform(df_features[saved_num_cols])
    
    # 5. Generate predictions
    return loaded_model.predict(df_features)

print("Preprocessing and inference function ready!")
```
